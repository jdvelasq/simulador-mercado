```python
CONFIG = load_config()
precios_10_1 = {}
precios_10_2 = {}
precios_10_3 = {}
precios_50_1 = {}
precios_50_2 = {}
precios_50_3 = {}
precios_90_1 = {}
precios_90_2 = {}
precios_90_3 = {}

for t_base in range(204, 241):
    
    precio_10_1 = []
    precio_10_2 = []
    precio_10_3 = []
    precio_50_1 = []
    precio_50_2 = []
    precio_50_3 = []
    precio_90_1 = []
    precio_90_2 = []
    precio_90_3 = []
    
    for escenario_ONI in CONFIG["ONI"]["ESCENARIOS"]:

        for escenario_Demanda in CONFIG["DEMANDA"]["ESCENARIOS"]:

            for i_serie in range(CONFIG["APORTES"]["NUMSIM"]):
                
                for horizonte in range(t_base, t_base + 24):

                    filename_precio = "{}/PRECIO_Agente_{:04d}_ONI_{}_dem_{}.csv".format(
                    t_base, i_serie, escenario_ONI, escenario_Demanda
                )
                    df_precio = pd.read_csv(filename_precio)

                    if escenario_ONI == 10 and escenario_Demanda == 1:

                        precio_10_1.append(df_precio.syntetic_Precio[horizonte])
                        
                    if escenario_ONI == 10 and escenario_Demanda == 2:

                        precio_10_2.append(df_precio.syntetic_Precio[horizonte])
                    
                    if escenario_ONI == 10 and escenario_Demanda == 3:

                        precio_10_3.append(df_precio.syntetic_Precio[horizonte])
                        
                    if escenario_ONI == 50 and escenario_Demanda == 1:

                        precio_50_1.append(df_precio.syntetic_Precio[horizonte])
                        
                    if escenario_ONI == 50 and escenario_Demanda == 2:

                        precio_50_2.append(df_precio.syntetic_Precio[horizonte])
                    
                    if escenario_ONI == 50 and escenario_Demanda == 3:

                        precio_50_3.append(df_precio.syntetic_Precio[horizonte])
                        
                    if escenario_ONI == 90 and escenario_Demanda == 1:

                        precio_90_1.append(df_precio.syntetic_Precio[horizonte])
                        
                    if escenario_ONI == 90 and escenario_Demanda == 2:

                        precio_90_2.append(df_precio.syntetic_Precio[horizonte])
                    
                    if escenario_ONI == 90 and escenario_Demanda == 3:

                        precio_90_3.append(df_precio.syntetic_Precio[horizonte])

                precios_10_1[t_base] = precio_10_1
                precios_10_2[t_base] = precio_10_2
                precios_10_3[t_base] = precio_10_3
                precios_50_1[t_base] = precio_50_1
                precios_50_2[t_base] = precio_50_2
                precios_50_3[t_base] = precio_50_3
                precios_90_1[t_base] = precio_90_1
                precios_90_2[t_base] = precio_90_2
                precios_90_3[t_base] = precio_90_3

                #df_precios_10_1 = pd.DataFrame(precios_10_1)
```
